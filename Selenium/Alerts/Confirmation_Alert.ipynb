```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

url = "https://www.w3schools.com/js/tryit.asp?filename=tryjs_confirm"
driver.get(url)

inframe_element = driver.find_element(By.ID, 'iframeResult')
driver.switch_to.frame(inframe_element)

button = driver.find_element(By.XPATH, '/html/body/button')
button.click()
time.sleep(1
          )
print(f'Alert text: {driver.switch_to.alert.text}')

# driver.switch_to.alert.accept()
# print("You clicked Accept..!")
# time.sleep(2)

driver.switch_to.alert.dismiss()
print("You clicked cancel..!")
time.sleep(2)

driver.switch_to.default_content()
time.sleep(1)

driver.quit()

```
