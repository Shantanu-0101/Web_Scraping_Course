```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome()

time.sleep(5)
driver.maximize_window()

url = "https://www.google.com"
driver.get(url)

print(driver.title)
print(driver.current_url)

#driver.save_screenshot("Google_Screenshot.png")
print("Screenshot Taken..!")

time.sleep(5)
driver.quit()
```
